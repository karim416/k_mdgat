```python
#get file object
filename="loop4.txt"
f = open(filename, "r")
epoch=[]
loss=[]
tloss=[]
gaploss=[]
t1=[]
t2=[]
t3=[]
while(True):
    line = f.readline()
    #if line is empty, you are done with all lines in the file
    if not line:
        break
    if line[0:5].strip() == 'epoch' :
        #you can access the line
        epoch.append(int(line.split()[2]))
        loss.append(float(line.split()[6]))
        tloss.append(float(line.split()[10]))
        gaploss.append(float(line.split()[14]))
    elif line[0:2].strip() == 'T1':
        t1.append(float(line.split()[3]))
        t2.append(float(line.split()[7]))
        t3.append(float(line.split()[11]))
#close file
f.close()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(22,6),facecolor='white')
plt.subplot(1,3,1)
plt.plot(epoch,loss)
plt.ylabel('loss totale',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(1,3,2)
plt.plot(epoch,tloss)
plt.ylabel('loss Transformation (R,t)',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


plt.subplot(1,3,3)
plt.plot(epoch,gaploss)
plt.ylabel('loss matching',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

fig.suptitle('Loss = f(epoch)',fontsize=22)

plt.savefig('loss.png', facecolor=fig.get_facecolor(), transparent=False)


plt.show()
```

```python
fig=plt.figure(figsize=(22,6),facecolor='white')
plt.subplot(1,3,1)
plt.plot(epoch,t1)
plt.ylabel('loss Transformation 1',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.subplot(1,3,2)
plt.plot(epoch,t2)
plt.ylabel('loss Transformation 2',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.subplot(1,3,3)
plt.plot(epoch,t3)
plt.ylabel('loss Transformation 3',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
fig.suptitle('Loss = f(epoch)',fontsize=22)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('tloss.png', facecolor=fig.get_facecolor(), transparent=False)


plt.show()
```

```python
fig=plt.figure(figsize=(18,6),facecolor='white')
plt.plot(epoch,t1,label='Trans 1')
plt.plot(epoch,t2,label='Trans 2')
plt.plot(epoch,t3,label='Trans 3')

plt.ylabel('loss Transformation',fontsize=18)
plt.xlabel('epoch',fontsize=18)
plt.grid()
plt.legend(fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.savefig('same_tloss.png', facecolor=fig.get_facecolor(), transparent=False)


plt.show()
```
